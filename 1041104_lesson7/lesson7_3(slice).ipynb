```python
cities = ['台北','台中','高雄','台南']
cities
```

```python
# 索引index
# 從第一個開始算起為 0
# 從最後一個開始算起為 -1
cities[-1]
```

```python
# slice[前面的索引:後面的索引]
# 前面的索引包含,後面的索引不包含
cities[0:1]
cities[0:2]
cities[:2]
cities[2:4]
cities[2:]
```

```python
# 前5個
greetings = list('Hello World')
greetings[:5]
```

```python
# 後5個
greetings[-5:]
```
