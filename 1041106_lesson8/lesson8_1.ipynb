```python
import os

def get_names(file_name):
    # 使用names.txt的絕對路徑
    current_dir = os.path.dirname(os.path.abspath("__file__"))
    file_path = os.path.join(current_dir,'assets',file_name)

    with open(file_path, encoding="utf-8") as file:
        content = file.read()

    return content.split('\n') # names.txt裡面是用\n
    
names = get_names("/workspaces/python_ai/1041106_lesson8/assets/names.txt")
len(names) #幾筆資料
```

```python
# dict的建立和取出元素值
# random.randint(a,b) 產生a到b的隨機整數
# info 指的是一個dict資料
# for in 迴圈列出學生名和隨機成績

# 名字有隨機，成績也有隨機
import random
students_names = random.sample(names,5) # 隨機抽5筆資料
for name in students_names: # for in 迴圈列出學生名
    info = {"姓名": name,  # information 資料
            "國文": random.randint(60,100), # 產生60~100的隨機整數
            "英文": random.randint(60,100),
            "數學": random.randint(60,100)
            }
    print(info) # 顯示info內容
```

```python
# dict的建立和取出元素值
# random.randint(a,b) 產生a到b的隨機整數
# info 指的是一個dict資料
# for in 迴圈列出學生名和隨機成績
# list+dict [] 存多筆dict資料[{},{}]
# scores 指的是一個list資料
# .append() 增加元素值到list中


import random
students_names = random.sample(names,5) # 隨機抽5筆資料
scores = [] # 建立一個空的list
for name in students_names: # for in 迴圈列出學生名
    info = {"姓名": name,
            "國文": random.randint(60,100), # 產生60~100的隨機整數
            "英文": random.randint(60,100),
            "數學": random.randint(60,100)
            }
    scores.append(info) # 把info增加到scores裡面

scores # 出現5筆dict資料
    
```

```python
# dict的建立和取出元素值
# random.randint(a,b) 產生a到b的隨機整數
# info 指的是一個dict資料
# for in 迴圈列出學生名和隨機成績
# list+dict [] 存多筆dict資料[{},{}]
# scores 指的是一個list資料
# .append() 增加元素值到list中
# def 自定function get_scores(names, num=5)

import random
def get_scores(names, num=10): # 定義function get_scores且預設抽10筆 names是參數，num是參數
    students_names = random.sample(names,num) # 隨機抽n筆資料
    scores = [] # 建立一個空的list
    for name in students_names: # for in 迴圈列出學生名
        info = {"姓名": name,
                "國文": random.randint(60,100), # 產生60~100的隨機整數
                "英文": random.randint(60,100),
                "數學": random.randint(60,100)
                }
        scores.append(info) # 把info增加到scores裡面
    return scores
student_scores = get_scores(names,num=5) # name是全域參數，num是參數，5是實際傳入的值
student_scores
    
```

```python
# 使用csv模組寫入CSV檔案


import csv

fieldnames = student_scores[0].keys() # 取得欄位名稱
current_dir = os.path.dirname(os.path.abspath("__file__")) # 取得目前目錄
file_path = os.path.join(current_dir,'1041106_lesson8/assets','test.csv') # 定義CSV檔案路徑

with open(file_path, 'w', newline='', encoding='utf-8') as csvfile: # 開啟CSV檔案
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames) # 建立DictWriter物件
    writer.writeheader() # 寫入欄位名稱
    for d in student_scores: # 逐筆寫入資料
        writer.writerow(d) # 寫入一筆資料
```
