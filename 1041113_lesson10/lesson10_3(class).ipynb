```python
# 自訂初始化方法 __init__
# self 代表實體本身
# 建立Person實體p1 且有name及age屬性

class Person:
    def __init__(self):
        self.name = "zoey"
        self.age = 38
```

```python
p1= Person()
print(p1.name)
print(p1.age)
```

```python
# 自訂初始化方法 __init__ 且可以傳入參數

class Person:
    def __init__(self,name:str,age:int):
        self.name = name
        self.age = age
```

```python
p1 = Person("zoey Lo",38)
print(p1.name)
print(p1.age)
```

```python
p2 = Person(age=25,name="John")
print(p2.name)
print(p2.age)
```
