```python
class Person:
    def __init__(self,name:str,age:int):
        self.name = name
        self.age = age
   #  定義一個description方法,用來描述這個人的名字及年齡     
    def description(self):
        print(f"我的名字叫:{self.name}")
        print(f"我的age是:{self.age}")
```

```python
# 建立Person實體p1 且有name及age屬性
# 呼叫description方法傳入參數name及age

p1 = Person(name="zoey Lo",age=38)
p1.description()
```

```python
# 繼承Person類別
# person是父類別
# student是子類別
# Student類別繼承Person類別的屬性及方法
class Student(Person):
    pass
```

```python
s1 = Student(name="zoey",age=38)
s1.description() #敘述式所以不用print
```

```python
class Student(Person):
    def __init__(self,name:str,age:int,chinese:int=60,english:int=60,math:int=60):
        super().__init__(name,age)
        self.chinese = chinese #前面的chinese是屬性，後面的chinese是參數
        self.english = english #前面的english是屬性，後面的english是參數
        self.math = math

    #覆寫(override)父類別的方法
    def description(self):
        super().description() #呼叫父類別的方法
        print(f"國文成績:{self.chinese}")
        print(f"英文成績:{self.english}")
        print(f"數學成績:{self.math}")
```

```python
s1 = Student(name="zoey",age=38,chinese=90,english=85,math=95) #有自訂成績
s1.description()
```

```python
# 如果手動 = 數字 將覆蓋掉原本的屬性值
s1.chinese = 100
s1.description()
```

```python
class Student(Person):
    @classmethod
    def nickname(cls): # nickname 是 class method
        print("我是class的method")

    @staticmethod 
    def heroname():
        print("我是static的method")
    
    def __init__(self,name:str,age:int,chinese:int=60,english:int=60,math:int=60):
        super().__init__(name,age)
        self.__chinese = chinese #__chinese 表示隱藏屬性(private)
        self.__english = english
        self.__math = math

    @property
    def chinese(self)->int:
        return self.__chinese
    
    @property
    def english(self)->int:
        return self.__english
    
    @property
    def math(self)->int:
        return self.__math
    
    #覆寫(override)
    def description(self):
        super().description()
        print(f"國文:{self.chinese}")
        print(f"英文:{self.english}")
        print(f"數學:{self.math}")
```
