```python
# 不管有沒有錯誤都會執行finally區塊的程式碼

try:
    money = int(input("請輸入金額")) 
except: 
    print("輸入錯誤")
else:
    print(money)
finally:
    print("應用程式結束")
```

```python
# 處理多種錯誤
# 增加應用程式的可靠度

# 輸入5000a時會出現格式錯誤
# 輸入-5000時會出現金額不可以小於0  


try:
    money = int(input("請輸入金額")) 
    if money <0:
        raise Exception ("金額不可以小於0")
except ValueError:
    print("格式錯誤")
except Exception as error: 
    print(error)
else:
    print(money)
finally:
    print("應用程式結束")
```
