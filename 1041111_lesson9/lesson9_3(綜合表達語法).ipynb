```python
# [1, 3, 5, 7, 9] 傳統方法建立
# range() 函數產生一個數列
# 使用 for 迴圈遍歷數列，並用 if 條件篩選奇數

numbers = []
for i in range(1,10):
    if i % 2 == 1:
        numbers.append(i) # .append() 方法將元素加入列表

numbers
```

```python
# 使用list的comprehension建立
# [運算式 for item in iterable]
# [運算式 for item in iterable if 條件]

print([i for i in range(1,10)])
print([i for i in range(1,10) if i % 2 == 1])
```

```python
word = 'HelloPython' # word為字串
word.count('l') # 計算字串中 'l' 出現的次數
```

```python
# dict的comprehension
word = 'letters'
print({letter:word.count(letter) for letter in word})
print({letter:word.count(letter) for letter in word if letter == 'e'})
```

```python
# type hinting 提示
numbers:list[int] = [1, 3, 5, 7, 9]
weeks:list[str] = ["星期一","星期二","星期三","星期四","星期五"]
```

```python
# 屬性檢查 type
print(type(weeks))
```

```python
# for in 迴圈 列出清單內容
for week in weeks:
    print(week)
```
